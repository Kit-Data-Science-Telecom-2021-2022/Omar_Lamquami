```python
def Preprocessing(df):
    df.columns = ['GDH', 'Rang', 'Voile', 'Skipper/Bateau', 'Heure', 'Lat', 'Long', 
         'Cap_30', 'Vitesse_30', 'VMG_30', 'Distance_30',
         'Cap_last', 'Vitesse_last', 'VMG_last', 'Distance_last',
         'Cap_24', 'Vitesse_24', 'VMG_24', 'Distance_24', 
         'DTF', 'DTL']
    df=df.drop(0)
    df = df.dropna(axis=1, how="all")
    
    #Nettoyge de la colonne voile/Nationalité
    df["Voile"]=df["Voile"].str.split('\n').apply(lambda x: x[-1])
    df.insert(3,"nat",df["Voile"].str.split(' ').apply(lambda x: x[0]))
    df["Voile"]=df["Voile"].str.split(' ').apply(lambda x: x[1])
    
    #Nettoayage de la colonne Skipper/Bateau
    df["Skipper/Bateau"].str.split("\n").apply(lambda x: x[0])
    df.insert(5,"Bateau",df["Skipper/Bateau"].str.split("\n").apply(lambda x: x[1]))
    df["Skipper/Bateau"]=df["Skipper/Bateau"].str.split("\n").apply(lambda x: x[0])
    df.rename(columns={'Skipper/Bateau':'Skipper'})
    
    #Formattage de la collone Heure
    date = re.search('([0-9]{8})', "20201108_140000.xlsx").group(0)
    df['Heure'] = df['Heure'].apply(lambda x: datetime.datetime.strptime(date+'_'+x[:5], '%Y%m%d_%H:%M') if  isinstance(x,str) else x)

    
    #convertir les colonnes numériques en données exploitables(sans unités)
    num_cols = list(df.columns[8:])
    df[num_cols] = df[num_cols].apply(lambda x: x.astype(str).str.replace(re.compile('[a-z]+|°'), '')).apply(pd.to_numeric)
    
    return df

    
df=pd.read_excel("20201108_140000.xlsx",header=[0,1],skiprows=3,nrows=33)
df=Preprocessing(df)

display(df)
```
